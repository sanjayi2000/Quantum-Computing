```python
import numpy as np
from qiskit import QuantumCircuit
import matplotlib.pyplot as plt
```

```python
# Step 1: Create a Quantum Circuit
qc = QuantumCircuit(3, 3)
```

```python
# Step 2: Apply gates
qc.ccx(0, 1, 2)  # Toffoli gate
qc.swap(1, 2)    # Swap gate
```

```python
# Step 3: Measure the qubits
qc.measure([0, 1, 2], [0, 1, 2])
```

```python
# Step 4: Draw the circuit using Matplotlib
qc.draw('mpl')  # This will display the circuit
```

```python
# Step 5: Simulate measurement outcomes
# We can simulate the outputs manually by using a random number generator
# For a 3-qubit circuit, we have 2^3 = 8 possible states (000, 001, ..., 111)
num_shots = 1024
possible_states = ['000', '001', '010', '011', '100', '101', '110', '111']
counts = {state: 0 for state in possible_states}
```

```python
# Simulate measurements
np.random.seed(42)  # For reproducibility
for _ in range(num_shots):
    # Randomly choose a state among the possible states
    chosen_state = np.random.choice(possible_states)
    counts[chosen_state] += 1
```

```python
# Step 6: Plot the measurement results using Matplotlib
labels = counts.keys()
values = counts.values()

plt.bar(labels, values)
plt.title('Simulated Measurement Results')
plt.xlabel('States')
plt.ylabel('Counts')
plt.xticks(rotation=45)
plt.show()
```
